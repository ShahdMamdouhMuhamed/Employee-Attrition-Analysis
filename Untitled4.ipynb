**Detecting and Handling Outliers**

```python

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

outlier_summary = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR


    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
    outlier_summary[col] = outliers.shape[0]

outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outlier_Count'])
outlier_df = outlier_df.sort_values(by='Outlier_Count', ascending=False)

print("Outlier Summary across all numerical columns:")
display(outlier_df)
```

**Descriptive Statistics**

```python


num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print(" Numerical Columns:\n")
for col in num_cols:
    print(f"--- {col} ---")
    print(f"Mean   : {df[col].mean():.2f}")
    print(f"Median : {df[col].median():.2f}")
    print(f"Mode   : {df[col].mode()[0]}")
    print("-" * 40)

print("\n Categorical Columns:\n")
for col in cat_cols:
    print(f"--- {col} ---")
    print(f"Most Frequent (Mode): {df[col].mode()[0]}")
    print(f"Unique Values Count : {df[col].nunique()}")
    print("-" * 40)
```

**(Correlation Heatmap)**

```python

numeric_cols = df.select_dtypes(include=['int64', 'float64'])


corr_matrix = numeric_cols.corr()

plt.figure(figsize=(15, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap for Numerical Features", fontsize=14)
plt.show()
```

```python

numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr_pearson = numeric_df.corr(method='pearson')

plt.figure(figsize=(18,10))
sns.heatmap(corr_pearson, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Pearson Correlation Heatmap (Numerical Columns Only)")
plt.show()
```

```python

numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr_spearman = numeric_df.corr(method='spearman')

plt.figure(figsize=(18,10))
sns.heatmap(corr_spearman, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Spearman Correlation Heatmap (Numerical Columns Only)")
plt.show()


```
